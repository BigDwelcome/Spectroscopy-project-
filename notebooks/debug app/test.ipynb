```python
import pandas as pd
from chemotools.derivative import SavitzkyGolay
from chemotools.scatter import StandardNormalVariate
import numpy as np
```

```python
df = pd.read_csv('2024-12-28T10-05_export.csv')
```

```python
# convertir les float en array 
df["spectre_converti"] = df["spectre"].apply(lambda x: [float(i) for i in x.split(",")])
```

```python
# Extraction des valeurs sous forme d'array numpy final

np.array(df['spectre_converti'].to_list())
```

```python
df['spectre_converti'].values
```

```python
snv = StandardNormalVariate()
first_derivative = SavitzkyGolay(15,2,1)
second_derivative = SavitzkyGolay(15,2,2)

# Appliquer les transformations sur toutes les lignes en une seule fois
spectre_array = np.array(df['spectre_converti'].tolist())

# Transformation SNV
df["spectre_snv"] = snv.fit_transform(spectre_array).tolist()

# Transformation première dérivée
df["spectre_first_derivative"] = first_derivative.fit_transform(spectre_array).tolist()

# Transformation seconde dérivée
df["spectre_second_derivative"] = second_derivative.fit_transform(spectre_array).tolist()
```
